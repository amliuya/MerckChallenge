```python

# Import packages
import numpy as np
import time
from sklearn.metrics import mean_squared_error
from sklearn import datasets
import xgboost as xgb

from hyperopt import hp, tpe, fmin, Trials, STATUS_OK
from collections import OrderedDict

reg_prob = datasets.make_friedman1(n_samples=100, n_features=10, noise=1.0, random_state=None)
x_train = reg_prob[0][0:50]
y_train = reg_prob[1][0:50]
x_test = reg_prob[0][50:100]
y_test = reg_prob[1][50:100]

space= hp.choice('booster', ['gbtree','gblinear','dart'])



def objective(booster):
    
    
    regressor = xgb.XGBRegressor(booster=booster)
    
#    max_depth = int(max_depth)
#    n_estimators = int(n_estimators)
    
    # Training 
    regressor.fit(x_train, y_train)
    
    # Making predictions and evaluating
    y_pred = regressor.predict(x_test)
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)
    

    return rmse

trials = Trials()
best = fmin(objective,
    space=space,
    algo=tpe.suggest,
    max_evals=100,
    trials=trials)

print(best)
print(trials.results)
```
